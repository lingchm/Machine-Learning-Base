# Image Compression

*SVD*

**Problem**

Given an image in the form of an array $X$, compress the image to different resolutions.

**Method**

We first perform Singular Value Decomposition and truncate to the first $r$ components.

$$
X = U \Sigma V^T
$$

$$
X' = U_r \Sigma_r V_r^T, r < m
$$

**References**

Steve Brunton: https://www.youtube.com/watch?v=H7qMMudo3e8&list=PLMrJAkhIeNNSVjnsviglFoY2nXildDCcv&index=6


```python
import numpy as np
import matplotlib.pyplot as plt
import os
from matplotlib.image import imread
plt.rcParams['figure.figsize'] = [16, 8]
```

```python
# read in an image
A = imread('data/cat.jpg')
X = np.mean(A, -1) # convert RGB to grayscale

img = plt.imshow(X)
img.set_cmap('gray')
plt.axis('off')
plt.show()
```

```python
U, S, VT = np.linalg.svd(X, full_matrices = False)
S = np.diag(S)
```

```python
# Construct approximate image using different number of singlar vectors
j = 0
for r in (5, 10, 20, 50, 100):
    X_ = U[:,:r] @ S[0:r,:r] @ VT[:r,:]
    plt.figure(j+1)
    j += 1
    img = plt.imshow(X_)
    img.set_cmap('gray')
    plt.axis('off')
    plt.title('r = ' + str(r))
    plt.show()
```

```python
# finding the best r 

# looking at singular values
plt.figure(1)
plt.semilogy(np.diag(S))
plt.title('Singular values')
plt.show() 

# looking at cumulative sum of singular values
plt.figure(2)
plt.plot(np.cumsum(np.diag(S)) / np.sum(np.diag(S)))
plt.title('Singular Values: Cumulative Sum')
plt.show()
```

**Discussion**

SVD provides a flexible way to represent the original matrix X by keeping the most important information up to a rank. The plots demonstrate that lot of 'energy' or information is captured in the first singular values. Even only keeping the 1st component we have 30% of the energy.
